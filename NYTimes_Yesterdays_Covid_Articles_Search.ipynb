```python
#NYTimes data Yesterdays data - USING ARTICLE SEARCH API

# load needed libraries
import requests 
import pandas as pd
import math 
import time
from datetime import date
import datetime
from datetime import timedelta

# find date of today and yesterday
today = date.today()
yesterday = today - timedelta(days = 1)

# split the date into year, month, day
df = pd.DataFrame(data = {'yesterday': yesterday}, index = {0})
df['year'] = pd.DatetimeIndex(df['yesterday']).year
df['month'] = pd.DatetimeIndex(df['yesterday']).month
df['day'] = pd.DatetimeIndex(df['yesterday']).day

# set month, day, year index 
month_index = df['month'][0]
day_index = df['day'][0]
year_index = str(df['year'][0])+'-'

# define my api key 
key = 'IFHrJfmGWoXsZn8If9WLj2Sfom9Zemyk'

# define an empty list for the Articles data
Articles = []

# create date string for the API url
if month_index < 10:
    if day_index < 10:
        date = year_index+'0'+str(month_index)+'-0'+str(day_index)
    else:
        date = year_index+'0'+str(month_index)+'-'+str(day_index)
else:
    if day_index < 10:
        date = year_index+str(month_index)+'-0'+str(day_index)
    else:
        date = year_index+str(month_index)+'-'+str(day_index)         

# sleep 7 seconds per NYTimes limit
time.sleep(7)
# define my url for calling the api for covid
url_covid = 'https://api.nytimes.com/svc/search/v2/articlesearch.json?q=("covid-19")&fq=pub_date:'
# call the api and save response into 'resp_covid' object
resp_covid = requests.get(url_covid+'('+date+')'+'&api-key='+key).json()
# get number of articles
articles_covid = resp_covid['response']['meta']['hits']
# sleep 7 seconds per NYTimes limit
time.sleep(7)
# define my url for calling the api for all
url_all = 'https://api.nytimes.com/svc/search/v2/articlesearch.json?fq=pub_date:'
# call the api and save response into 'resp_all' object
resp_all = requests.get(url_all+'('+date+')'+'&api-key='+key).json()
# get number of articles
articles_all = resp_all['response']['meta']['hits']
# convert to a dataframe for date i
Articles_i = [date, articles_covid, articles_all]
# append to list, of lists of Articles data
Articles.append(Articles_i)
    
# create df
NYTimes_Yesterday = pd.DataFrame(Articles ,columns=['Date', 'Amount_of_covid_Articles', 'Amount_of_All_Articles'])
NYTimes_Yesterday['%_of_covid_Articles'] = round((NYTimes_Yesterday['Amount_of_covid_Articles']/NYTimes_Yesterday['Amount_of_All_Articles'])*100,2)
# print df
NYTimes_Yesterday
```
