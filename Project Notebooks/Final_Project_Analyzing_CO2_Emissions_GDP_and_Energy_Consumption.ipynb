### Merging all the data from the British Petroleum Source into One Data Frame

Now that we have all the data that we need in well-organized data frames, we can merge them into a unique data frame.  
The keys for the merge are:  
- Country
- Year 

```python
# merge the seperate data frames for the types of Renewable Energy types 
# into one data frame
merge=CO2Emiss_df_m.merge(Geo_BioEnrgCons_df_m,  how='left', left_on=['Country','year'], right_on=['Country','year']).merge(WindEnrgCons_df_m, how='left', left_on=['Country','year'], right_on=['Country','year']).merge(SolarEnrgCons_df_m, how='left', left_on=['Country','year'], right_on=['Country','year']).merge(HydroEnrgCons_df_m, how='left', left_on=['Country','year'], right_on=['Country','year']).merge(PrimaryEnrgCons_df_m,how='left', left_on=['Country','year'], right_on=['Country','year'])

# view the first few rows of the new merged data frame  
merge.head()
```

## Merging the Data from our three sources, to create a unique dataset for our Project
___

First, we check the names of the countries in the data from all three sources to see if they match and the data can be merged, or if we have to replace some names if they don't match before we merge the data.   

```python
# print the names of the Countries in our three data frames

print('GDP Country name values:',GDP_melt['Country'].unique())

print('Population Country name values:',pop_melted['Country'].unique())

print('Consumption Country name values:',countries_scope_consumption['Country'].unique())
```

It's possible to see that there are several different versions of names for: Russia, China and OECD.   
We are going to replace the name values that don't match with consistent names.

```python
# replace names that don't match with a consistent name value 
GDP_melt['Country'].replace({'OECD members':'OECD','Russian Federation':'Russia'},inplace=True)
# check values
GDP_melt['Country'].unique()
```

```python
# replace names that don't match with a consistent name value 
pop_melted['Country'].replace({'OECD - Total':'OECD',"China (People's Republic of)":'China'},inplace=True)
# check values
pop_melted['Country'].unique()
```

```python
# replace names that don't match with a consistent name value 
countries_scope_consumption['Country'].replace({'Russian Federation':'Russia','of which: OECD':'OECD'},inplace=True)
# check values
countries_scope_consumption['Country'].unique()
```

Now we can merge the three sources.   
First merge GDP and Energy Consumption data.

```python
# merge the seperate data frames GDP and Energy Consumption
# into one data frame
cons_GDP=pd.merge(GDP_melt[['Country','year','GDP_million']],countries_scope_consumption,left_on=['Country','year'], right_on=['Country','year'],how='outer')

# check country values to be sure merge was successful
cons_GDP.Country.unique()
```

Now, merge with population data, to have all three sources.   
We used the outer join because we want to have all the possible data present in the three data sources.

```python
# merge the data frame containing GDP and Energy Consumption with the Population data frame  
final_df=pd.merge(cons_GDP,pop_melted[['Country','year','Population_million']],left_on=['Country','year'], right_on=['Country','year'],how='outer')
```

## Research Question 1
___
In the past five years, what is the percentage of renewable energy consumption of the total energy consumption in the BRICS nations in comparison to the OECD nations?

To answer the first question:  
- We are going to subset the data frame keeping only the total yearly consumption of Energy value and the three different renewable sources of energy. 
- We will also subset the data frame keeping only the last 5 years.
- We will create a new column with the total renewable energy consumption. 
- We will create a new column with the proportion of renewable energy consumption from total energy consumption. 
- We will plot the proportion using a side by side bar plot for each of the BRICS nations and OECD.

```python
# create copy of the final data set for Q1, dropping the columns we don't need 
q1_df=final_df.drop(columns=['Population_million','GDP_million','Mtoe_CO2Emis']).copy()

# subset the df with values only for the last 5 years (2014-2018)
q1_df=q1_df.iloc[:,:][q1_df['year'] > 2013].copy()

# create new column for Total Renewable Energy Consumption,
# by combining the consumption values of all the renewable types 
q1_df['Total_renewable']=q1_df['Mtoe_Geo']+q1_df['Mtoe_Wind']+q1_df['Mtoe_Solar']+q1_df['Mtoe_Hydro']

# create new column for the percentage of Renewable Energy Consumption out of Total Energy Consumption 
q1_df['Percentage_renewable']=(q1_df['Total_renewable']/q1_df['Mtoe_Total'])*100

# set double index of Contry and year
q1_df.set_index(['Country','year'], inplace = True)

# view the head of the data frame
q1_df.head()
```

```python
# create one figure for all the subplots
fig_q1, ax = plt.subplots(figsize=(15,7))

# make side-by-side bar plot for the percentage of Renewable Energy Consumption out of Total Energy Consumption, 
# for each Country in our scope
q1_df.groupby(['Country','year']).sum()['Percentage_renewable'].unstack('Country').plot(kind='bar',ax=ax);

# set yaxis label, and plot title
ax.yaxis.set_label_text("Percentage");
ax.set_title("Percentage of Renewable Energy Consumption out of Total Energy Consumption");
# make xticks horizontal 
plt.xticks(rotation=0);

plt.close();
```

```python
fig_q1
```

We can conclude the following from the above chart:
- Out of the BRICS countries, Brazil is leading in the use of renewable energy. Most of Brazil's renewable energy is coming from Hydro Energy. It seems that Brazil is growing sustainably.
- South Africa is the country that consumes the least renewable energy out of the BRICS countries.
- China has a higher renewable energy consumption percentage than the OECD. 

## Research Question 2
___
Looking at the yearly trend (from 1990 to 2018), what is the relationship between the GDP (per capita) and Total Energy Consumption (per capita) in the BRICS nations?

To answer the second question:  
- We will create a subset with the trend for each BRICS nation with two columns: GDP (per capita) and Total energy consumption (per capita). Using trend is important considering the possible models that we can perform. In fact GDP and Energy consumption are usually in literature considered autocorrelated variables. This means that use the trend could prevent this problem.  
- We will create a scatter plot for each BRICS nation to find the relationship between the two variables.
- We will also insert the regression line to evaluate the relationship.

```python
# create copy of the final data set for Q2, dropping the columns we don't need 
q2_df=final_df[['Country','year','Mtoe_Total','Population_million','GDP_million']].copy()

# create the two new columns that we need: Total consumption per capita 
# and GDP per capita
q2_df['Mtoe_Total_percapita']=q2_df['Mtoe_Total']/q2_df['Population_million']
q2_df['GDP_percapita']=q2_df['GDP_million']/q2_df['Population_million']

# view the head of the data frame
q2_df.head()
```

```python
import numpy as np
# create another subset of the data frame with only the columns that we need for the plot
q2_df_res2=q2_df[['Country','year','Mtoe_Total_percapita','GDP_percapita']].copy()

# we set and sort the index. This is necessary to create the trend values.
q2_df_res2.set_index(['Country','year'],inplace=True)
q2_df_res2.sort_index()

# create the two tren columns as null values
q2_df_res2['toe_per_capita_diff'] = np.nan
q2_df_res2['GDP_percapita_diff'] = np.nan

# we have to save the indexes
idx = pd.IndexSlice

# using a loop we create the values for the two variables.
for ix in q2_df_res2.index.levels[0]:
    q2_df_res2.loc[ idx[ix,:], 'toe_per_capita_diff'] = q2_df_res2.loc[idx[ix,:], 'Mtoe_Total_percapita' ].diff()
    q2_df_res2.loc[ idx[ix,:], 'GDP_percapita_diff'] = q2_df_res2.loc[idx[ix,:], 'GDP_percapita' ].diff()

# now we have to resort the index
q2_df_res2.sort_index()
```

```python
fig_q2, axes= plt.subplots(figsize=(15,10))

# since we have 5 countries in BRICS and the OECD we need 6 plots
# we decided to have them in 2 rows and 3 columns
# looking at trend data:
# for each plot we want to see the relationship between Total Consuption per capita and GDP per capita.

# moreover, in the title we insert the correlation between the two variables

plt.subplot(2, 3, 1)
fig_br = sns.regplot( y=q2_df_res2['toe_per_capita_diff']['Brazil'], x=q2_df_res2['GDP_percapita_diff']['Brazil'],data=q2_df_res2,color="c")
fig_br.set_title('Brazil :%s' %str(round(q2_df_res2['toe_per_capita_diff']['Brazil'].corr(q2_df_res2['GDP_percapita_diff']['Brazil']),2)))

plt.subplot(2, 3, 2)
fig_chi=sns.regplot( y=q2_df_res2['toe_per_capita_diff']['China'], x=q2_df_res2['GDP_percapita_diff']['China'],data=q2_df_res2,color="c")
fig_chi.set_title('China :%s' %str(round(q2_df_res2['toe_per_capita_diff']['China'].corr(q2_df_res2['GDP_percapita_diff']['China']),2)))

plt.subplot(2, 3, 3)
fig_ind=sns.regplot( y=q2_df_res2['toe_per_capita_diff']['India'], x=q2_df_res2['GDP_percapita_diff']['India'],data=q2_df_res2,color="c")
fig_ind.set_title('India:%s' %str(round(q2_df_res2['toe_per_capita_diff']['India'].corr(q2_df_res2['GDP_percapita_diff']['India']),2)))

plt.subplot(2, 3, 4)
fig_SA=sns.regplot( y=q2_df_res2['toe_per_capita_diff']['South Africa'], x=q2_df_res2['GDP_percapita_diff']['South Africa'],data=q2_df_res2,color="c")
fig_SA.set_title('South Africa :%s' %str(round(q2_df_res2['toe_per_capita_diff']['South Africa'].corr(q2_df_res2['GDP_percapita_diff']['South Africa']),2)))

plt.subplot(2, 3, 5)
fig_RUS=sns.regplot( y=q2_df_res2['toe_per_capita_diff']['Russia'], x=q2_df_res2['GDP_percapita_diff']['Russia'],data=q2_df_res2,color="c")
fig_RUS.set_title('Russia :%s' %str(round(q2_df_res2['toe_per_capita_diff']['Russia'].corr(q2_df_res2['GDP_percapita_diff']['Russia']),2)))

plt.subplot(2, 3, 6)
fig_OECD=sns.regplot( y=q2_df_res2['toe_per_capita_diff']['OECD'], x=q2_df_res2['GDP_percapita_diff']['OECD'],data=q2_df_res2,color="c")
fig_OECD.set_title('OECD :%s' %str(round(q2_df_res2['toe_per_capita_diff']['OECD'].corr(q2_df_res2['GDP_percapita_diff']['OECD']),2)))

plt.suptitle('Correlation between Consumption and GDP')

plt.subplots_adjust(hspace=0.3)
plt.subplots_adjust(wspace=0.3)


plt.close()
```

```python
fig_q2
```

We can see that considering the trend data, all the countries have a positive correlation between GDP and Total Energy Consumption, as can be expected.  

Furthermore, it is also interesting to look at the relationship between the yearly values of the variables (not the yearly differences), to relate it to the Kuznets curve theorem.  

```python
# create another subset of the data frame with only the columns that we need for the plot
q2_df_res=q2_df[['Country','year','Mtoe_Total_percapita','GDP_percapita']].copy()

# we set and sort the index
q2_df_res.set_index(['Country','year'],inplace=True)

# create the figure containing all the plots
fig_q2, axes= plt.subplots(figsize=(15,10))

# since we have 5 countries in BRICS and the OECD we need 6 plots
# we decided to have them in 2 rows and 3 columns
# looking at yearly data:
# for each plot we want to see the relationship between Total Consuption per capita and GDP per capita.

# moreover, in the title we insert the correlation between the two variables
plt.subplot(2, 3, 1)
fig_br = sns.regplot( y=q2_df_res['Mtoe_Total_percapita']['Brazil'], x=q2_df_res['GDP_percapita']['Brazil'],data=q2_df_res,color="c")
fig_br.set_title('Brazil :%s' %str(round(q2_df_res['Mtoe_Total_percapita']['Brazil'].corr(q2_df_res['GDP_percapita']['Brazil']),2)))

plt.subplot(2, 3, 2)
fig_chi=sns.regplot( y=q2_df_res['Mtoe_Total_percapita']['China'], x=q2_df_res['GDP_percapita']['China'],data=q2_df_res,color="c")
fig_chi.set_title('China :%s' %str(round(q2_df_res['Mtoe_Total_percapita']['Brazil'].corr(q2_df_res['GDP_percapita']['Brazil']),2)))

plt.subplot(2, 3, 3)
fig_ind=sns.regplot( y=q2_df_res['Mtoe_Total_percapita']['India'], x=q2_df_res['GDP_percapita']['India'],data=q2_df_res,color="c")
fig_ind.set_title('India:%s' %str(round(q2_df_res['Mtoe_Total_percapita']['India'].corr(q2_df_res['GDP_percapita']['India']),2)))

plt.subplot(2, 3, 4)
fig_SA=sns.regplot( y=q2_df_res['Mtoe_Total_percapita']['South Africa'], x=q2_df_res['GDP_percapita']['South Africa'],data=q2_df_res,color="c")
fig_SA.set_title('South Africa :%s' %str(round(q2_df_res['Mtoe_Total_percapita']['South Africa'].corr(q2_df_res['GDP_percapita']['South Africa']),2)))

plt.subplot(2, 3, 5)
fig_RUS=sns.regplot( y=q2_df_res['Mtoe_Total_percapita']['Russia'], x=q2_df_res['GDP_percapita']['Russia'],data=q2_df_res,color="c")
fig_RUS.set_title('Russia :%s' %str(round(q2_df_res['Mtoe_Total_percapita']['Russia'].corr(q2_df_res['GDP_percapita']['Russia']),2)))

plt.subplot(2, 3, 6)
fig_OECD=sns.regplot( y=q2_df_res['Mtoe_Total_percapita']['OECD'], x=q2_df_res['GDP_percapita']['OECD'],data=q2_df_res,color="c")
fig_OECD.set_title('OECD :%s' %str(round(q2_df_res['Mtoe_Total_percapita']['OECD'].corr(q2_df_res['GDP_percapita']['OECD']),2)))

plt.suptitle('Correlation between Consumption and GDP')
plt.show()
```

## Research Question 3
___
What was the yearly CO2 emission level (per capita) in each of the BRICS nations between 1990 and 2018 compared to the OECD average? And what the average CO2 emission level in the BRICS nations compared to the OECD average?

To answer the third question:  
- We are going to create a subset containing the population and CO2 emission data for the BRICS nations and OECD.
- We will create a new column for per capita carbon dioxide emissions. 
- We will create a total CO2 emission level for the BRICS nations (required grouping). To create it, we will need to  add a new column to the data frame stating OECD or BRICS nation.
- We will visualize the data using two different line plots. First showing each of the BRICS nations separately, then showing a comparison between the OECD and the BRICS nations as a united entity.


```python
# now we subset again the dataset to mantain only the columns that we need for the plot
q3_df=final_df[['Country','year','Mtoe_CO2Emis','Population_million']].copy()

# now we need to create a new columns that will tell if a country is part of OECD or BRICS to group later
q3_df['country_group']=['BRICS'  if (x=='China' or x=='India' or x=='Brazil' or x=='South Africa' or x=='Russia') else 'OECD' for x in q3_df['Country']]

# we also need to create the CO2 emission per capita column
q3_df['Mtoe_CO2Emis_per_capita']= q3_df['Mtoe_CO2Emis']/q3_df['Population_million']

# small sample
q3_df.head()
```

```python
# now we want to show two plots:
    #1) shows all the countries in our scope, over time
    #2) shows only the two grouped categories

# since there will be two different charts we need to set two different axes,
# that would help in calling the right chart in the right position
fig_q3, (ax1, ax2) = plt.subplots(2, 1,figsize=(15,7))

# we want the chart with all the countries to be first, so we put ax1
fig_q3_1= q3_df.groupby(['Country','year']).mean()['Mtoe_CO2Emis_per_capita'].unstack('Country').plot(ax=ax1);
fig_q3_1.set_title ('CO2 Emissions (Per Capita) In BRICS Over Time');
ax1.legend(bbox_to_anchor=(1.0, 1.05))

# we want the chart with the two groups to be second, so we put ax2
fig_q3_2 = q3_df.groupby(['country_group','year']).mean()['Mtoe_CO2Emis_per_capita'].unstack('country_group').plot(ax=ax2);
fig_q3_2.set_title('Avg CO2 Emissions (Per Capita) In BRICS and OECD');
ax2.legend(bbox_to_anchor=(1.1, 1.05))

plt.subplots_adjust(hspace=0.4)
plt.close();
```

```python
fig_q3
```

We can see that the CO2 emission level (per capita) has increased only in China between 1990 and 2018. Compared to the OECD average, Russia has a higher level of CO2 Emissions since 2010. Brazil and India have remained at a steady low rate of CO2 Emissions during the time period. 
Comparing the average CO2 Emission level in the BRICS nations to the OECD average, we can see that it is rising while the OECD average is dropping.
